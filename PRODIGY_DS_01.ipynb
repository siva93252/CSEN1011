```python
import matplotlib.pyplot as plt
import numpy as np

# Generating example dataset of ages
np.random.seed(0)
ages = np.random.randint(18, 80, 1000)  # Generating 1000 random ages between 18 and 80

# Creating histogram
plt.hist(ages, bins=20, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages in a Population')

# Showing the plot
plt.show()













```
