```python
 import pandas as pd
import numpy as np
import folium
from folium.plugins import HeatMap

# Create a sample traffic accident dataset
data = {
    'Crash Date': pd.date_range(start='1/1/2023', end='1/31/2023', freq='D'),
    'Municipality Name': np.random.choice(['Town1', 'Town2', 'Town3', 'Town4', 'Town5'], size=31),
    'Latitude': np.random.uniform(40.5, 41, size=31),
    'Longitude': np.random.uniform(-74.5, -73.5, size=31),
    'Total Killed': np.random.randint(0, 5, size=31),
    'Cell Phone In Use Flag': np.random.choice(['N', 'Y'], size=31)
}

df1 = pd.DataFrame(data)

# Create a list of unique town names from df1 for use later to make a call to function to get Lat, Long
town_names = df1['Municipality Name'].unique()

# Define a base map generator function
def generateBaseMap(default_location=[40.7128, -74.0060], default_zoom_start=10):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map

# Create a base map
base_map = generateBaseMap()

# Add markers to the map
for i, row in df1.iterrows():
    popup_text = f"Total Killed: {row['Total Killed']}"
    folium.Marker([row['Latitude'], row['Longitude']], popup=popup_text).add_to(base_map)

# Display the map
base_map
```
